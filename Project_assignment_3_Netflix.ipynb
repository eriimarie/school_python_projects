```python
#Q1. Make a line chart that compares the number of titles of each release year of 5 countries of your choice.
#You can regard show_id as the unique identifier of a title.

#Some records have a NaN value in the country column.
#Make sure to remove such records from the dataset first 
#(see https://stackoverflow.com/questions/13413590/how-to-drop-rows-of-pandas-dataframe-whose-value-in-a-certain-column-is-nan (Links to an external site.)). 
#Also note that value in the country column is a string that may contain multiple countries.
#Consider using the str.contains() method (see https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.contains.html (Links to an external site.)) 
#and the .loc method (see https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html (Links to an external site.)) 
#to extract records that include your target country in the string. The
#result should look like this:

import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('netflix-titles.csv') 
a=[]
b=[]
c=[]
d=[]
e=[]
year=[]
for i in range(1980,2020):
    x=0
    y=0
    h=0
    k=0
    l=0
    for j in range(0,5000):
        if df['release_year'][j]==i and df['country'][j]=="Korea": 
            x=x+1;
        if df['release_year'][j]==i and df['country'][j]=="Japan":
            y=y+1;
        if df['release_year'][j]==i and df['country'][j]=="China":
            h=h+1;
        if df['release_year'][j]==i and df['country'][j]=="Taiwan":
            k=k+1;
        if df['release_year'][j]==i and df['country'][j]=="HongKong":
            l=l+1;
    
    a.append(x)
    b.append(y)
    c.append(h)
    d.append(k)
    e.append(l)
    year.append(i)

#plot lines
plt.title('Comparison Chart of number of titles per year')
plt.xlabel('Release Year', fontsize=14)
plt.ylabel('Number of Titles', fontsize=14)
plt.plot(year, a, label = "Korea") #label based on the country selected
plt.plot(year, b, label = "Japan") ##label based on the country selected
plt.plot(year, c, label = "China")
plt.plot(year, d, label = "Taiwan")
plt.plot(year, e, label = "Hongkong")
plt.legend()
plt.show()
```
