```python
#Question 12

#Write a program that counts the number of odd numbers, even numbers,
#and squares from 2 to 25 (inclusive).

import math

#intialize count starting with 0
oddCnt = 0
evenCnt = 0
sqrCnt = 0

#count the between the number 2,26 0f 0dd and even
for x in range(2, 26): 
    if x%2 == 0: #check for even numbers
        evenCnt += 1 #increasing counts by 1
    elif x%2 == 1: #else check for odd number 
        oddCnt += 1   #increasing counts by 1 
    if math.sqrt(x) in range(2,26):  #checking for the range (2,26) squared root
        sqrCnt += 1 #increasing counts by 1
        
#Print the counts
print("Count of odd numbers: ", oddCnt)
print("Count of even numbers: ", evenCnt)
print("count of squares: ", sqrCnt)
```

```python
#question 14 

#Rewrite this following “for” loop as a “while” loop:

   
for i in range(0, 17):
    if i % 2 == 0:
        print(i)

#Rewrite program: 
i = 0  #initialize i at 0
while i<17:  #you are giving range 17
    if i%2 == 0: #checking if it is an even number
        print(i)
    i += 1 #increment i    
```

```python
#question 15
#Simplify the following code so that it does not use any breaks or continue statements, 
#yet the logic is equivalent.


counter = 0
while True:
    if counter >= 100:
        break
    elif counter % 2 == 0:
        counter += 1
        continue
    else:
            counter *= 2;
            continue
            
#Simplify form: 

counter = 0 #initilize count start with 0

while counter < 100:  #while the loop count the number until 100
    if counter % 2 == 0: #count with remainder of 0
        counter += 1 #increase count by 1
    else:
            counter *= 2 #multiple count by 2
```

```python
#question 35
#"Prompts" here means prompting the user with input(). 
#Let's say the user enters 500, you should store 500 in 
#a variable that holds the sum, and prompt the user again. 
#Now the user enters the second time with 400, 
#add 400 to the sum variable, prompt the user again 
#(because the sum is still below 1000). 
#Now the user enters the third time with 300, 
#your loop should stop because the sum is now 1200.

#Of course the numbers above are randomly chosen. 
#The key point of this question is, 
#we don't know what number the user will enter each time,
#and hence we don't know how many times to prompt. 
#A while loop would be useful here. 
#Why not a for loop? 
#for loop handles cases where we already know how many times a loop should execute. e.g.
#for i in range(0, 9): #loop 9 times
#for char in "python": #loop 6 times


#Write a program that continually prompts for positive integers and 
#stops when the sum of numbers entered exceeds 1000.
#Stop early if a negative integer is entered.

sum = 0 #initialize sum starting with 0

while sum < 1000: #while loop sum to 1000
    x = int(input("Please enter a Positive Integer: ")) #asking the user to enter a positive integer
    if x < 0: #if x less than of greater that 0
        break  #break here
    else: #else print the sum plus x
        sum = sum + x

print("Sum is: ", sum) #print the result and stop when exceeds to 1000
```
