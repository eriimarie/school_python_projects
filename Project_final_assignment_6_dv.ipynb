```python
#3.At least two visualizations - at least one for statistical summary, at least one for analysis.
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('world_happiness_report_2020.csv')

# Filtering df for ratings with 5 stars.
df = df[df['Healthy life expectancy'] >70]

# Grouping by brands
df = df.groupby('Country name').count()

# Filtering brands rated 5 star for atleast 5 times.  
df1 = df[df['Healthy life expectancy'] >=2]

df.reset_index(level=0, inplace=True)
plt.pie(df['Healthy life expectancy'], labels=df['Country name'])
plt.show()
```
